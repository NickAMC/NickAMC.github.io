# Support Vector Machines

## Maximal Margin Classifier

* Goal: Find the optimal separating hyperplane that produces the largest margin between two classes of data points in a feature space.

* Margin: The distance between the separating hyperplane and the closest data points from each class. (These closest points are called support vectors).

* Why "Maximal"?: The idea is that a larger margin leads to better generalization of the classifier to new, unseen data.

## Key Points about the Maximal Margin Classifier:

* Linearly Separable Case: It works best when the data is linearly separable (meaning a clean line or hyperplane can divide the classes perfectly).

* Overfitting Dangers: In cases where data is not perfectly linearly separable, the maximal margin classifier can be overly sensitive to individual data points, potentially leading to overfitting.


## Support Vector Machines (SVMs)

Support Vector Machines build upon the Support Vector Classifier concept and provide a powerful tool for both linear and non-linear classification tasks.

* Kernel Trick: SVMs employ the "kernel trick," which involves mapping data into a higher-dimensional space where linear separation may be possible. This allows for the creation of non-linear decision boundaries in the original feature space.

* Common Kernels: Popular choices include:

     * Linear Kernel (for linearly separable cases)

     * Polynomial Kernel

     * Radial Basis Function (RBF) Kernel

## Key Points about SVMs

* Robust and Versatile: SVMs are resilient to noise and are a highly effective classification method for various datasets.

* Hyperparameter Tuning: Selecting the right kernel and tuning parameters like 'C' is crucial for optimal SVM performance.

```python
import pandas as pd 
import numpy as  np 
import matplotlib.pyplot as plt 
```

```python
# create SVM class

class SVM() :

     def __init__(self, learning_rate = 0.001, lambda_parm = 0.01, n_iters = 1000) -> None:
          self.lr = learning_rate
          self.lambda_param = lambda_parm
          self.n_iters = n_iters
          self.w = None
          self.b = None
     
     def  fit(self, X,y) : 
          n_samples, n_features = X.shape

          y_ = np.where(y<=0, -1,1)

          # init weights
          self.w = np.zeros(n_features)
          self.b = 0

          for _ in range(self.n_iters):
               for idx, x_i in enumerate(X):
                    condition = y_[idx] * (np.dot(x_i, self.w) - self.b) >= 1
                    if condition :
                         self.w -= self.lr * (2 * self.lambda_param * self.w)
                    else: 
                         self.w -= self.lr * (2* self.lambda_param * self.w - np.dot(x_i, y_[idx]))
                         self.b -= self.lr * y_[idx]

     def predict(self,X) :
          approx = np.dot(X, self.w) - self.b
          return np.sign(approx)
```

```python
if __name__ == "__main__":
     from sklearn.model_selection import train_test_split
     from sklearn import datasets

     X,y= datasets.make_blobs(n_samples=50, n_features=2, centers =2, cluster_std=1.05, random_state=40)

     y = np.where(y==0,-1,1)

     X_train,X_test, y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=123)
```

```python
clf = SVM()
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
```

```python
def accuracy(y_true, y_pred) :
     accuracy = np.sum(y_true == y_pred) / len(y_true)
     return accuracy
```

```python
print(f'SVM accuracy: {accuracy(y_test, predictions)}')
```

```python
def visualize_svm():
     def get_hyperplane_value(x, w, b, offset):
          return (-w[0] * x + b + offset) / w[1]

     fig = plt.figure()
     ax = fig.add_subplot(1, 1, 1)
     plt.scatter(X[:, 0], X[:, 1], marker="o", c=y)

     x0_1 = np.amin(X[:, 0])
     x0_2 = np.amax(X[:, 0])

     x1_1 = get_hyperplane_value(x0_1, clf.w, clf.b, 0)
     x1_2 = get_hyperplane_value(x0_2, clf.w, clf.b, 0)

     x1_1_m = get_hyperplane_value(x0_1, clf.w, clf.b, -1)
     x1_2_m = get_hyperplane_value(x0_2, clf.w, clf.b, -1)

     x1_1_p = get_hyperplane_value(x0_1, clf.w, clf.b, 1)
     x1_2_p = get_hyperplane_value(x0_2, clf.w, clf.b, 1)

     ax.plot([x0_1, x0_2], [x1_1, x1_2], "y--")
     ax.plot([x0_1, x0_2], [x1_1_m, x1_2_m], "k")
     ax.plot([x0_1, x0_2], [x1_1_p, x1_2_p], "k")

     x1_min = np.amin(X[:, 1])
     x1_max = np.amax(X[:, 1])
     ax.set_ylim([x1_min - 3, x1_max + 3])

     plt.show()

visualize_svm()
```
